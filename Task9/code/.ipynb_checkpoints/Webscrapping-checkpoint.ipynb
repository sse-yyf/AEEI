```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

## 9.1

```python
path_save = "../outcome"

url = 'https://www.numbeo.com/cost-of-living/'
try:
    res = requests.get(url)
    res.raise_for_status()
except Exception as exc:
    print('There was a problem: %s' % (exc))

html_file_path = os.path.join(path_save, 'numbeo_cost_of_living.html')
with open(html_file_path, 'w', encoding='utf-8') as html_file:
    html_file.write(res.text)
```

## 9.2

```python
soup = BeautifulSoup(res.content, 'html.parser')

# Find countries
countries = []
country_options = soup.find('select', {'id': 'country'})
for option in country_options('option')[1:]:
    country = option.text.strip()
    countries.append(country)
```

## 9.3

```python
# save one country file

country = countries[99] # randomly pick one

url = f'https://www.numbeo.com/cost-of-living/country_result.jsp?country={country.replace(" ","-")}&displayCurrency=USD'

try:
    res = requests.get(url)
    res.raise_for_status()
except Exception as exc:
    print('There was a problem: %s' % (exc))
    
html_file_path = path_save + '/'+ str(country) + '.html'
with open(html_file_path, 'w', encoding='utf-8') as html_file:
    html_file.write(res.text)
    
# load beautiful soup
soup = BeautifulSoup(res.content, 'html.parser')

table = soup.find('table', {'class':'data_wide_table new_bar_table'})
data = {}
rows = table.find_all('tr')
for row in rows:
    price = []
    isdiv = row.find('div')
    if isdiv:
        continue
    else:
        columns = row.find_all('td')
        category = columns[0].text.strip()
        price.append(columns[1].text.strip().replace("\xa0",""))
        data[category]= price

div = soup.find('div', {'class': "seeding-call table_color summary limit_size_ad_right padding_lower other_highlight_color"})
lis = div.find_all('li',{'class':''})
cost = []
rent = []
cost.append(lis[0].text.strip())
rent.append(lis[1].text.strip())
data["Cost of living compared to Sweden"] = cost
data["Rent compared to Sweden"] = rent

df = pd.DataFrame(data).T
df = df.rename(columns = {0:'Price'})
df.to_csv(path_save + '/'+ str(country) + '.csv')

```

## 9.4

```python
path_save_csv = "../outcome/csv"
path_save_html = "../outcome/html"

for country in countries:
    url = f'https://www.numbeo.com/cost-of-living/country_result.jsp?country={country.replace(" ","-")}&displayCurrency=USD'

    try:
        res = requests.get(url)
        res.raise_for_status()
    except Exception as exc:
        print('There was a problem: %s' % (exc))

    html_file_path = path_save_html + '/'+ str(country) + '.html'
    with open(html_file_path, 'w', encoding='utf-8') as html_file:
        html_file.write(res.text)

    # load beautiful soup
    soup = BeautifulSoup(res.content, 'html.parser')

    table = soup.find('table', {'class':'data_wide_table new_bar_table'})
    
    if table:
        data = {}
        rows = table.find_all('tr')
        for row in rows:
            price = []
            isdiv = row.find('div')
            if isdiv:
                continue
            else:
                columns = row.find_all('td')
                category = columns[0].text.strip()
                price.append(columns[1].text.strip().replace("\xa0",""))
                data[category]= price

        div = soup.find('div', {'class': "seeding-call table_color summary limit_size_ad_right padding_lower other_highlight_color"})
        if div:
            lis = div.find_all('li',{'class':''})
            cost = []
            rent = []
            cost.append(lis[0].text.strip())
            rent.append(lis[1].text.strip())
            data["Cost of living compared to Sweden"] = cost
            data["Rent compared to Sweden"] = rent

            df = pd.DataFrame(data).T
            df = df.rename(columns = {0:'Price'})
            df.to_csv(path_save_csv + '/'+ str(country) + '.csv')
        else:
            print(country + ':no info compared to Sweden')
    else:
        print(country + ':no table')
```
